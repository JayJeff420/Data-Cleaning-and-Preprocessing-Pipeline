```python

import pandas as pd
from sklearn.preprocessing import StandardScaler
url = 'https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv'
df = pd.read_csv(url)
print(df.head())
print(df.info())
print(df.describe(include='all'))
print("Missing values before handling:")
print(df.isnull().sum())
df['Age'].fillna(df['Age'].mean(), inplace=True)
print("Missing values after handling:")
print(df.isnull().sum())
duplicates = df.duplicated().sum()
print(f'Duplicate records: {duplicates}')
df.drop_duplicates(inplace=True)
duplicates = df.duplicated().sum()
print(f'Duplicate records after removal: {duplicates}')
print("Unique values in 'Age':", df['Age'].unique())
df['FamilySize'] = df['Siblings/Spouses Aboard'] + df['Parents/Children Aboard'] + 1
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
df.drop(columns=['Name'], inplace=True)
num_features = ['Age', 'Fare', 'FamilySize']
scaler = StandardScaler()
df[num_features] = scaler.fit_transform(df[num_features])
print(df.info())
print(df.head())
df.to_csv('cleaned_titanic_data.csv', index=False)
```
